```python
df2 = df[df['decommissioned'] == 0]
```

```python
df3 = df2[['zip', 'primary_city', 'state', 'country', 'timezone', 'latitude', 'longitude']]
```

```python
df4 = df3.rename(columns={'zip':'zipcode', 'primary_city':'city'})
```

```python
df4
```

```python
df4['zipcode'] = df4['zipcode'].astype(str)
```

```python
df4['zipcode'] = df4['zipcode'].str.pad(width=5, side='left', fillchar='0')
```

```python
df4.to_csv('preprocessed_zipcode.csv', index=False)
```
